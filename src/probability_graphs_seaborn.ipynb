```python
import pandas as pd
import seaborn as sns

data = pd.read_csv("data_random.csv")
```

Seaborn'la Probability Grafikleri

```python
sns.set_style('whitegrid')
sns.displot(data, x="saat", col="Y", kde = False, color ='red', bins = 30)
sns.displot(data, x="haftanin gunu", col="Y", kde = False, color ='red', bins = 30)
sns.displot(data, x="haftaici / sonu", col="Y", kde = False, color ='red', bins = 30)
sns.displot(data, x="ay", col="Y", kde = False, color ='red', bins = 30)
sns.displot(data, x="mevsim", col="Y", kde = False, color ='red', bins = 30)
```
